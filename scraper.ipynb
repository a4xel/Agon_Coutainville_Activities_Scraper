```python
""""Do the necessary imports"""
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
"""Scrape a first time to find the max number of pages in the agenda"""

url = "https://www.agoncoutainville.fr/agenda/"

response=requests.get(url)

soup_1 = BeautifulSoup(response.content, 'html.parser')

page_number= int(soup_1.find_all("a",class_="page-numbers")[1].get("href")[-2:]) 
```

```python
"""Scrape the date, title, and descriptive of all event from all the pages """

descriptive = []
title = []
date = []

for i in range(1,page_number+1):
    
    req = requests.get(url+ f"?page={i}")
    soup = BeautifulSoup(req.text, 'html.parser')
    
    for j in soup.find_all(class_="post-title"):
        title.append(j.text)
    
    for k in soup.find_all(class_="post-content"):
        descriptive.append(k.text)
        
    for l in soup.find_all("article"):
        date.append(l.find(class_="label-date-round top-align").text)  
```

```python
"""Arrange the date format when the event spans between two dates to make it more readable"""

date_formated=[]
for i in date:
    if len(i) > 15:
        i=i.replace("2022","2022/",1)
        date_formated.append(i)
    else:
        date_formated.append(i)
```

```python
"""Create final list of tuples with the zip function. Convert the list to a Dataframe for better readability"""

result = zip(date_formated,title,descriptive)
tableau_des_activités = pd.DataFrame(list(result), columns = ['Date', "Nom de l'activité",'Description'])
```

```python
"""Extract the obtained dataframe to csv format"""

file_path = os.getcwd()

tableau_des_activités.to_csv(file_path+"/activities.csv")
```
